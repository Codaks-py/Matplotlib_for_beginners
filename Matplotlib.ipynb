```python
# Matplotlib 
# drawing a pie chart from a random generated values
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

series = pd.Series(3 * np.random.rand(4), 
                   index = ['A', 'B', 'C', 'D'],
                   name = 'series')

print(series)
# making use of the imported library 
plt.figure()
series.plot.pie(figsize=(3,3), subplots=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.rand(10,4),
            columns= ['A', 'B', 'C', 'D'])

print(df)
ax = df.plot.scatter(x='A', y='B',
                    color="Blue",
                    label='Group 1')
plt.figure()
df.plot.scatter(x="C", y='D',
               color="Green",
               label = "Group 2",
               )

plt.show()
```

```python
# from the above scatter graph we can as well make the two graph in just one graph like this
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.rand(10,4),
            columns= ['A', 'B', 'C', 'D'])

print(df)
ax = df.plot.scatter(x='A', y='B',
                    color="Blue",
                    label='Group 1')
plt.figure()
df.plot.scatter(x="C", y='D',
               color="Green",
               label = "Group 2",
               ax=ax)

plt.show()
# the only difference is we just telling the program to make both graphs in a page using the ax=ax i.e
# we assigned the first plot to ax right then when trying to plot the second one we didn't really assign it but we later did at the end which
# is why we later used ax=ax.
```

```python
#Hexgonal bin plot  #Hexbin
# Just like the scatter plot but in hexagonal shape
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

s = pd.DataFrame(np.random.rand(150,2),
                columns = ['A', 'B'])

s['B'] += np.arange(150)
print(s)

plt.figure()
s.plot.hexbin(x="A", y="B", gridsize = 20)
plt.show()
```

```python
# Histogram plot
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

s = pd.DataFrame({ 'a' : np.random.rand(1000) + 1,
                  'b' : np.random.rand(1000),
                  'c' : np.random.rand(1000) - 1},
                columns = ['a', 'b', 'c'])
print(s)
plt.figure()
s['a'].plot.hist(orientation = 'vertical', cumulative=True)
plt.show()
```

```python
import numpy as np
import pandas as pd
s = pd.Series([3,4,1,2])
s.cumsum()
# it operations is thus, index 0 takes 3, index 1 takes (3+4 = 7), index 2 takes (7+1 = 8)
# then index 3 takes (8+2 = 10), thats how it works.
# .cumsum() function calculates the cummulative sum of the values in the series
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

Df = pd.DataFrame(np.random.randn(100,2), 
                 columns =['B', 'C']).cumsum()
Df['A']= pd.Series(list(range(100)))
# cumsum() function is used to calculate the sum of the values 
print(Df)

plt.figure()
Df.plot(x='A', y='B')
plt.show()
```

```python
# imagine plotting multiply bar plot in 3D
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from mpl_toolkits import mplot3d

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# let define the color
colors = ['r', 'g', 'b', 'c','m','y','k']
ylicks = [0,1,2,3,4,5,6]
for c,k in zip(colors, ylicks):
    x = np.arange(25)
    y = np.random.rand(25)
    # making a bar plot
    ax.bar(x,y,zs=k, zdir='y', color=c, alpha=0.8)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d

fig = plt.figure()
ax = plt.axes(projection ='3d')

x = np.linspace(-10, 10, 30)
y = np.linspace(-10, 10, 30)

X,Y = np.meshgrid(x,y)
Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.contour(X,Y,Z, 50, cnap='coolwarm')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
n=3
data = np.linspace(0, 2*n*np.pi, 300)

def sinusoidal(sty):
    plt.style.use(sty)
    fig, ax = plt.subplots()
    ax.plot(data, np.sin(data), label='Sine')
    ax.plot(data, np.cos(data), label='Cosine')
    ax.legend()
    
    
sinusoidal('default')
plt.show()
```

```python
sinusoidal('seaborn')
plt.show()
```
